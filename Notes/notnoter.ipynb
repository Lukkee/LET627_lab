```python
import ipywidgets as widgets
from IPython.display import display, HTML

freq_slider = widgets.FloatSlider(
    value=200,
    min=20,
    max=4000,
    step=10,
    description='Frequency (Hz):',
    style={'description_width': '150px'}
)
output = widgets.Output()
def update_period(change):
    freq = change['new']
    period_sec = 1 / (2 * freq)
    period_us = period_sec * 1e6
    with output:
        output.clear_output()
        print(f"Frequency: {freq} Hz")
        print(f"Period: {period_us:.2f} μs")
freq_slider.observe(update_period, names='value')
update_period({'new': freq_slider.value})
display(freq_slider)
display(output)
```
