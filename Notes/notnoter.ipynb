### Metaforiska, bokstavliga och kanske lite filosofiska

````python
import ipywidgets as widgets
from IPython.display import display
def calculate_period(freq):
    """Calculate and display period for given frequency"""
    period_sec = 1 / (2 * freq)
    period_us = period_sec * 1e6
    print(f"Frequency: {freq} Hz")
    print(f"Period: {period_us:.2f} μs")
freq_slider = widgets.interactive(
    calculate_period,
    freq=widgets.FloatSlider(
        value=200,
        min=20,
        max=4000,
        step=10,
        description='Frequency (Hz):',
        style={'description_width': '150px'}
    )
)
display(freq_slider)
````

## Frequency indices

````python
import ipywidgets as widgets
from IPython.display import display

notes = ['A', 'A#', 'B', 'C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#']

def calculate_frequency(k):
    p0 = 440.0  # base freq
    fk = p0 * (2 ** (k / 12))
    pk = (1 / (2 * fk)) * 1e6
    print(f"frequency p({k}) = {fk:.4f} Hz")
    print(f"period p({k}) = {pk:.4f} μs")
    print("note: " + notes[k % 12])

index_slider = widgets.interactive(
    calculate_frequency,
    k=widgets.IntSlider(
        value=0,
        min=-12,
        max=12,
        step=1,
        description='Index:',
        style={'description_width': '150px'}
    )
)
display(index_slider)
````

````python
import ipywidgets as widgets
from IPython.display import display

notes = ['A', 'A#', 'B', 'C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#']
indices = [0, 2, 4, 0, 0, 2, 4, 0, 4, 5, 7, 4, 5, 7, 7, 9, 7, 5, 4, 0, 7, 9, 7, 5, 4, 0, 0, -5, 0, 0, -5, 0]


def calculate_indices(key):
    tuned_indices = [value + key for value in indices]
    print(f"indices: " + str(tuned_indices))
    print(f"min: " + str(min(tuned_indices)))
    print(f"max: " + str(max(tuned_indices)))


key_slider = widgets.interactive(
    calculate_indices,
    key=widgets.IntSlider(
        value=0,
        min=-5,
        max=5,
        step=1,
        description='Key:',
        style={'description_width': '150px'}
    )
)

display(key_slider)
````
