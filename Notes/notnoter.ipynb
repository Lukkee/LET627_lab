```python
import ipywidgets as widgets
from IPython.display import display
def calculate_period(freq):
    """Calculate and display period for given frequency"""
    period_sec = 1 / (2 * freq)
    period_us = period_sec * 1e6
    print(f"Frequency: {freq} Hz")
    print(f"Period: {period_us:.2f} μs")
freq_slider = widgets.interactive(
    calculate_period,
    freq=widgets.FloatSlider(
        value=200,
        min=20,
        max=4000,
        step=10,
        description='Frequency (Hz):',
        style={'description_width': '150px'}
    )
)
display(freq_slider)
```

```python
import ipywidgets as widgets
from IPython.display import display

notes = ['A', 'A#', 'B', 'C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#']

def calculate_frequency(k):
    p0 = 440.0  # base freq
    pk = p0 * (2 ** (k / 12))
    print(f"p({k}) = {pk:.4f} Hz")
    print("note: " + notes[k % 12])

index_slider = widgets.interactive(
    calculate_frequency,
    k=widgets.IntSlider(
        value=0,
        min=0,
        max=12,
        step=1,
        description='Index:',
        style={'description_width': '150px'}
    )
)
display(index_slider)
```
